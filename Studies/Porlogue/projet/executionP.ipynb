```python
from pyswip import Prolog

# Charger SWI-Prolog
prolog = Prolog()
prolog.consult("coloriage.pl")  # Charger le fichier Prolog

# Exécuter la requête Prolog
resultats = list(prolog.query("map_coloring([A, B, C, D, E])"))

# Afficher les solutions
for solution in resultats:
    print(f"Solution trouvée : {solution}")
```
